```python
# Titanic Dataset - EDA Notebook
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset('titanic')

# Data overview
print("First 5 rows:")
print(df.head())

print("\nDataset shape:")
print(df.shape)

print("\nIndex:")
print(df.index)

print("\nColumns:")
print(df.columns)

print("\nData types:")
print(df.dtypes)

print("\nMissing values per column:")
print(df.isnull().sum())

print("\nSummary statistics:")
print(df.describe())

# --------------------
# Data Filtering Examples
# --------------------

print("\nPassengers who survived:")
print(df[df['survived'] == 1].head())

print("\nFemales under age 30:")
print(df[(df['age'] < 30) & (df['sex'] == 'female')].head())

# --------------------
# Value Counts
# --------------------

print("\nPassenger class distribution:")
print(df['class'].value_counts())

print("\nSurvival distribution:")
print(df['survived'].value_counts())

# --------------------
# Visualizations
# --------------------

sns.countplot(x='survived', data=df)
plt.title("Survival Count")
plt.xlabel("Survived")
plt.ylabel("Count")
plt.show()

sns.histplot(data=df, x='age', hue='survived', bins=30)
plt.title("Age Distribution by Survival")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```
